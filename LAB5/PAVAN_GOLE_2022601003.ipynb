```python
# Load necessary libraries
install.packages(c("ggplot2", "dplyr", "wordcloud", "RColorBrewer", "plotly", "car"))
library(ggplot2)
library(dplyr)
library(wordcloud)
library(RColorBrewer)
library(plotly)
library(car)
```

```python
data <- read.csv("Housing.csv")
```

```python
# Box Plot
# Box plot of price by furnishing status
ggplot(data, aes(x=furnishingstatus, y=price, fill=furnishingstatus)) +
  geom_boxplot() +
  labs(title="Box Plot of Price by Furnishing Status", x="Furnishing Status", y="Price") +
  theme_minimal()
```

**This data suggests that furnishing status does have some relationship with property prices, with furnished properties generally commanding higher prices, but there's significant variability within each category**

```python
# Violin Plot
# Violin plot of price by furnishing status
ggplot(data, aes(x=furnishingstatus, y=price, fill=furnishingstatus)) +
  geom_violin() +
  labs(title="Violin Plot of Price by Furnishing Status", x="Furnishing Status", y="Price") +
  theme_minimal()
```

**This data suggests that furnishing status does have some relationship with property prices, with furnished properties generally commanding higher prices, but there's significant variability within each category**

```python
# Linear Regression Plot
# Linear regression of price vs. area
ggplot(data, aes(x=area, y=price)) +
  geom_point(aes(color="Data Points")) +
  geom_smooth(method="lm", se=FALSE, color="blue") +
  labs(title="Linear Regression of Price vs. Area", x="Area", y="Price") +
  theme_minimal()
```

**area is a significant factor in determining property prices, other factors also play important roles, as evidenced by the significant variation around the regression line.**

```python
# Non-Linear Regression Plot
# Fit a quadratic model for price vs. area
ggplot(data, aes(x=area, y=price)) +
  geom_point(aes(color="Data Points")) +
  geom_smooth(method="lm", formula=y ~ poly(x, 2), se=FALSE, color="red") +
  labs(title="Non-Linear Regression (Quadratic) of Price vs. Area", x="Area", y="Price") +
  theme_minimal()
```

** the relationship between property area and price is more complex than a simple linear increase, with the rate of price increase changing (and potentially decreasing) for very large properties.**

```python
# 3D Chart
plot_ly(data, x = ~area, y = ~bedrooms, z = ~price,
        type = "scatter3d", mode = "markers",
        marker = list(size = 5, color = ~price, colorscale = "Viridis")) %>%
  layout(scene = list(xaxis = list(title = "Area"),
                      yaxis = list(title = "Bedrooms"),
                      zaxis = list(title = "Price")))
use gg plot
```

**If you have more area and more bedrooms  that area price will be high**

```python
# Jitter Plot
# Jitter plot of price by furnishing status
ggplot(data, aes(x=furnishingstatus, y=price, color=furnishingstatus)) +
  geom_jitter(width=0.2, height=0) +
  labs(title="Jitter Plot of Price by Furnishing Status", x="Furnishing Status", y="Price") +
  theme_minimal()
```

**It suggests that while furnishing status does influence price to some extent, it's likely not the sole or even primary determinant of property prices.**
